# T2* percentage error

```python

chosedNum = [11,10,4,5,3,7,9]
tr = 6
fa = 90
N = 7
echonum = 7


###Gold standard
#print(str(tr)+"_"+str(fa))
GOLDSTD_mean = np.asarray([], dtype = float)
GOLDSTD_std = np.asarray([], dtype = float)

GOLDstd_T2Star = np.load("MASK_4RESULT/GOLDstandardT2Star"+str(tr)+"TR_"+str(fa)+"FA_T2StarValue_"+str(echonum)+".npy")
NPeriod_T2Star =  np.load("MASK_4RESULT/NPeriodicT2Star"+str(tr)+"TR_"+str(N)+"_Period_T2StarValue.npy")


BA_plot, ax = plt.subplots(1,figsize = (13,5))

###Gold standard
#print(str(tr)+"_"+str(fa))
GOLDSTD_mean = np.asarray([], dtype = float)
GOLDSTD_std = np.asarray([], dtype = float)
GOLDSTD_data = np.asarray([], dtype = float)
i = 0

i = i+1
###N Periodic 
N_Period_data = np.asarray([], dtype = float)
N_Period_mean = np.asarray([], dtype = float)
N_Period_std = np.asarray([], dtype = float)

for phantomNum in chosedNum:
    GOLDSTD_data = np.append(GOLDSTD_data, GOLDstd_T2Star[phantomNum-1,:])
    GOLDSTD_mean = np.append(GOLDSTD_mean, np.mean(GOLDstd_T2Star[phantomNum-1,:]))
    GOLDSTD_std = np.append(GOLDSTD_std, np.std(GOLDstd_T2Star[phantomNum-1,:]))
 
    N_Period_data = np.append(N_Period_data, NPeriod_T2Star[phantomNum-1,:])
    N_Period_mean = np.append(N_Period_mean, np.mean(NPeriod_T2Star[phantomNum-1,:]))
    N_Period_std = np.append(N_Period_std, np.std(NPeriod_T2Star[phantomNum-1,:]))


diff = (GOLDSTD_data - N_Period_data)/GOLDSTD_data*100       
ax.scatter(GOLDSTD_data*1000, diff, alpha = 0.5)
ax.axhline(0, color='green', linestyle='--', alpha = 0.5)
ax.axhline(-10, color='red', linestyle='--', alpha =0.5)
ax.axhline(10, color='red', linestyle='--', alpha = 0.5)

#md, sd, mean, CI_low, CI_high = bland_altman_plot(GOLDSTD_data, N_Period_data, ax_BA)
#ax_BA.set_title(str(P)+" Period "+ str(tr)+"ms TR", fontweight='bold')
#plt.title(str(tr)+"ms TR "+str(N)+" Period", fontweight='bold',fontsize = 20)
plt.xlabel("Gold-Standard Measured T2* (ms)",fontsize = 15)
plt.ylabel("ka-SPGR T2* Percentage Bias (%)",fontsize = 15)

errorstd = np.asarray([], dtype = float)
errormean = np.asarray([], dtype = float)
for group in range (0,7):
    start = group * 109
    end = start + 109
    errorstd = np.append(errorstd, np.std(diff[start:end]))
    errormean = np.append(errormean, np.mean(diff[start:end]))
plt.errorbar(GOLDSTD_mean*1000, errormean, yerr=errorstd, fmt='ro', capsize=4)

plt.ylim(-25,25)
# plt.xlim(0,0.045)


#BA_plot.savefig("Figure7_1_7Period.jpg")
```

# Effective T2* SNR

```python
chosedNum = [11,10,4,5,3,7,9]
chosedNum_axis = [str(11),str(10),str(4),str(5),str(3),str(7),str(9)]
Period = [7,12]
echo = [7,12]
GOLDstd_time = 26
bar_width = 0.35

# N = 7
# NPeriod_time = 41

N = 12
NPeriod_time = 48

tr = 6
fa = 20
dummyTime = 5000*tr*1e-3
plt.figure(1)

axis_label = np.asarray([], dtype = str)
echonum = N


GOLDstd_T2Star = np.load("MASK_4RESULT/GOLDstandardT2Star"+str(tr)+"TR_"+str(fa)+"FA_T2StarValue_"+str(echonum)+".npy")
NPeriod_T2Star =  np.load("MASK_4RESULT/NPeriodicT2Star"+str(tr)+"TR_"+str(N)+"_Period_T2StarValue.npy")

GOLDSTD_mean = np.asarray([], dtype = float)
GOLDSTD_std = np.asarray([], dtype = float)

NPeriod_mean = np.asarray([], dtype = float)
NPeriod_std = np.asarray([], dtype = float)

GOLDSTD_SNR = np.asarray([], dtype = float)
NPeriod_SNR = np.asarray([], dtype = float)

for phantomNum in chosedNum:
    GOLD_mean = np.mean(GOLDstd_T2Star[phantomNum-1,:])
    GOLD_std = np.std(GOLDstd_T2Star[phantomNum-1,:])
    GOLDSTD_SNR = np.append(GOLDSTD_SNR, GOLD_mean/GOLD_std/np.sqrt(GOLDstd_time))

    NPeriod_mean = np.mean(NPeriod_T2Star[phantomNum-1,:])
    NPeriod_std = np.std(NPeriod_T2Star[phantomNum-1,:])
    NPeriod_SNR = np.append(NPeriod_SNR, NPeriod_mean/NPeriod_std/np.sqrt(NPeriod_time-dummyTime))
    axis_label = np.append(axis_label, "{:.2f}".format(GOLD_mean*1000))

print(GOLDSTD_SNR)
print(NPeriod_SNR)

# Set the position of each bar on the x-axis

r1 = range(len(chosedNum_axis))
r2 = [x + bar_width for x in r1]

plt.figure(figsize = (10,10))
# Plot the bars for Group 1
plt.bar(r1, GOLDSTD_SNR, width=bar_width, label='Gold standard')

# Plot the bars for Group 2
plt.bar(r2, NPeriod_SNR, width=bar_width, label=str(N)+' Periodic ka-SPGR')

# Add labels and title
plt.xlabel('Mean Ground Truth T2* of the phantom sphere (ms)',fontsize = 20)
plt.ylabel('Effective T2* Signal-to-Noise Ratio',fontsize = 20)
#plt.title(str(N)+' Period')
plt.ylim(0,15)

# Add x-axis tick labels
plt.xticks([r + bar_width/2 for r in range(len(chosedNum_axis))], axis_label, fontsize = 12)
plt.yticks(fontsize = 12)
# Add a legend
plt.legend(fontsize = 12)

# Show the plot
    #plt.show()
 

        
#plt.savefig("Figure8_12period.jpg")
```

```python
#N Period Data
# Define colormap
cmap = plt.cm.jet

chosedNum = [11,10,4,5,3,7,9]
#chosedNum = [11,10,4,5,3,7,9,13,14,12]

tr = 6
Period = 5
fa = 10
GOLDSTD_data = np.asarray([], dtype = float)
STANDARD_MASK_LABEL = np.load("MASK_3RESULT/MASKLABEL_4.npy")


kaSPGR_fitDATA = np.asarray([], dtype = float)
kaSPGR_fittime = np.asarray([], dtype = float)
for i in range (1,6):
    DICOM = 'recon/n_periodic/tr'+str(tr)+'_n'+str(Period)+'_echo'+str(i)+'.dcm'
    img = pydicom.read_file(DICOM)
    kaSPGR_DATA = img.pixel_array.reshape((256,256))
    kaSPGR_fitDATA = np.append(kaSPGR_fitDATA, kaSPGR_DATA[109,111])
    kaSPGR_fittime = np.append(kaSPGR_fittime, tr/2+tr*(i-1))

kaSPGR_fitDATA = kaSPGR_fitDATA/np.max(kaSPGR_fitDATA)
Fstate = np.asarray([0,1,2,3,4], dtype = int)
plt.figure(1)
xtick_positions = [0,1,2,3,4]
xtick_labels = ['F0','F1','F2','F3','F4']
plt.scatter(Fstate, kaSPGR_fitDATA)
plt.xticks(xtick_positions, xtick_labels)
plt.xlabel("F-states", fontsize = 15)
plt.ylabel("Normalised F-states magnitude", fontsize = 15)

    
plt.figure(2)
xtick_positions = kaSPGR_fittime
xtick_labels = ['TE = 3ms','TE + TR = 9ms','TE + 2TR = 15ms','TE + 3TR = 21ms','TE + 4TR = 27ms']
plt.scatter(kaSPGR_fittime,kaSPGR_fitDATA)
plt.xticks(xtick_positions, xtick_labels, rotation = -20)
plt.xlabel("time (ms)", fontsize = 15)
plt.ylabel("Normalised F-states magnitude", fontsize = 15)

def func(t, A, R2Star):
    return A * np.exp(-R2Star * t)
        #fitted coefficient stored in popt:[A R2*] --> T2* = 1/R2*
popt, pcov = curve_fit(func, kaSPGR_fittime, kaSPGR_fitDATA)
PixelR2Star = popt[1]
print(1/PixelR2Star)
```

```python
#GOlD Standard Data
# Define colormap
cmap = plt.cm.jet

chosedNum = [11,10,4,5,3,7,9]
#chosedNum = [11,10,4,5,3,7,9,13,14,12]

tr = 6
#gold standard
fa = 90
te1 = 3

GOLDSTD_fitDATA = np.asarray([], dtype = float)
GOLDSTD_fittime = np.asarray([], dtype = float)
for i in range (1,6):
    DICOM = 'recon/multiecho_FLASH/tr'+str(tr)+'_fa'+str(fa)+'_echo'+str(i)+'.dcm'
    img = pydicom.read_file(DICOM)
    GOLDSTD_DATA = img.pixel_array.reshape((256,256))
    GOLDSTD_fitDATA = np.append(GOLDSTD_fitDATA, GOLDSTD_DATA[109,111])
    GOLDSTD_fittime = np.append(GOLDSTD_fittime, te1+tr*(i-1))

GOLDSTD_fitDATA = GOLDSTD_fitDATA/np.max(GOLDSTD_fitDATA)
    
xtick_positions = GOLDSTD_fittime
xtick_labels = ['TE1 = 3ms','TE2 = 9ms','TE3 = 15ms','TE4 = 21ms','TE5 = 27ms']
plt.xticks(xtick_positions, xtick_labels, rotation = -20)
plt.scatter(GOLDSTD_fittime, GOLDSTD_fitDATA)
plt.xlabel("time (ms)", fontsize = 15)
plt.ylabel("Normalised F-states magnitude", fontsize = 15)

def func(t, A, R2Star):
    return A * np.exp(-R2Star * t)
        #fitted coefficient stored in popt:[A R2*] --> T2* = 1/R2*
popt, pcov = curve_fit(func, GOLDSTD_fittime, GOLDSTD_fitDATA)
PixelR2Star = popt[1]
print(1/PixelR2Star)
```
