# Bloch Simulation - bSSFP #

```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
# the parameters used in this simulation 
#time unit --- second
proton_density = 1.0
T1 = 1.0 
T2 = 100e-3
TR = 10e-3
TE = 5e-3
# tip angle alpha 10, 30, 90 degree
tip_angle = [10,30,90]
#initial magnetisation 
M0 = np.asarray([[0],[0],[1]],dtype = float)
TR_Num = 1000 #Repeat the excitation 1000 times
```

```python
# FUNCTIONS
#RF pulse excitation matrix
def RF_pulse(alpha, phi):
    #The RF pulse tip the spin "alpha" degree from z axis
    #and a phase shift of "phi" degree clockwise from x axis
    
    #convert in to radians 
    a = math.radians(alpha)
    p = math.radians(phi)
    #each term of the matrix
    r11 = math.cos(a)*math.sin(p)**2 + math.cos(p)**2
    r12 = (1 - math.cos(a))*math.cos(p)*math.sin(p)
    r13 = - math.sin(a)*math.sin(p)
    r21 = (1 - math.cos(a))*math.cos(p)*math.sin(p)
    r22 = math.cos(a)*math.cos(p)**2 + math.sin(p)**2
    r23 = math.sin(a)*math.cos(p)
    r31 = math.sin(a)*math.sin(p)
    r32 = - math.sin(a)*math.cos(p)
    r33 = math.cos(a)
    R = np.asarray([[r11,r12,r13],[r21,r22,r23],[r31,r32,r33]], dtype = float)
    
    return R

#T1 recovery, T2 decay terms effect for time t
def Relaxation(t, T1, T2):
    e2 = np.exp(-t/T2)
    e1 = np.exp(-t/T1)
    E = np.asarray([[e2 , 0, 0],[0, e2, 0],[0, 0, e1]], dtype = float)
    
    return E

#Precession term effect on off-resonance (f_off Hz) spin at time t
def Precession(f_off, t):
    #free-precession angle beta
    b = f_off*t*2*math.pi
    #rad
    #b = math.radians(beta)
    P = np.asarray([[math.cos(b), math.sin(b), 0],[-math.sin(b), math.cos(b), 0],[0, 0, 1]], dtype = float)
    
    return P
    
```

## PART 1: 
<p>For balance Steady State Free Precession with <b>no phase cycling ($\Delta\phi = 0$)</b></p>
<p>Modelling the signal for spins with <b>off-resonance = 0</b></p>
<p>Aim:</p> 
<p>See how transverse magnetisation magnitude ($|M_{xy}|$) changes as more TR performed (Number of TR = 1000).</p>
<p>(Want to see the transverse magnetisation magnitude reaching a steady state at about 100TRs)</p>

```python
f_off = 0
phase_cycle = 0
T1 = 1.0
T2 = 0.1
tip_angle = [10, 30, 90]

#----------------------------------------------------------------------------------------------
#draw for different tip angle, 10, 30, 90
plot_number = 1
for a in tip_angle:

    M_pre = M0 #initial pre pulse magnetisation 
    #store value
    TR_num = np.asarray([], dtype = int)
    M_xy = np.asarray([], dtype = float)

    for k in range(1,TR_Num+1):
    #repeat 1000 times
        #Tip
        M_post = RF_pulse(a,phase_cycle) @ M_pre
        #Consider T1 recovery, T2 decay and precession
        #@ time TE 
        P_TE = Precession(f_off,TE) #precession matrix at t = TE
        E_TE = Relaxation(TE, T1, T2) #Relaxation matrix at t = TE

        M_TE = P_TE @ E_TE @ M_post + (np.identity(3) - E_TE) @ M0 #magnetisation at TE
        M_transverse = np.sqrt(M_TE[0]**2+M_TE[1]**2)
        #store value in the array
        TR_num = np.append(TR_num, k)
        M_xy = np.append(M_xy, M_transverse)
        
        #print the final steady value
        if (k==1000):
            print("The steady state value at tip angle = "+str(a)+" is: "+str(M_transverse))


        #@ time TR 
        P_TR = Precession(0,TR) #precession matrix at t = TR
        E_TR = Relaxation(TR, T1, T2) #Relaxation matrix at t = TR
        M_TR = P_TR @ E_TR @ M_post + (np.identity(3) - E_TR) @ M0 #magnetisation at TR
        M_pre = M_TR

    #plot M_xy against number of TR
    plt.figure(plot_number)
    plt.plot(TR_num, M_xy)
    plt.xlabel("Number of TR")
    plt.ylabel("$|M_{xy}|$")
    plt.title("tip angle:" + str(a))
    plt.show()
    
    #save figure
    #plt.savefig("PartONE_" +str(int(T1*1000))+"_"+str(int(T2*1000))+"_tip_angle_"+ str(a) + ".jpg")
    #Part1_T1/T2_"+str(int(T1*1000))+"/"+str(int(T2*1000))+"
    plot_number = plot_number+1
    
```

<p>A larger tip angle has a smaller steady-state transverse magnetisation magnitude </p>
<p>- If it tips a larger angle, it needs more time to recover. However we use the same TR, so the magnitude will become smaller and smaller for each tip. For the 90° tip, it is almost zero</p>

## PART 2:
<p>For balance Steady State Free Precession with <b>no phase cycling ($\Delta\phi = 0$)</b></p>
<p>Modeling the signals from spins with different off-resonance, <b>off-resonance range (-400 ~ 400 Hz)</b></p>
<p>Aim:</p> 
<p>1. <b>Draw off-resonance profile</b> for different tip angle, 10, 30, 90 degree. At steady state, see how different off-resonance signal have different steady state transverse magnetisation magnitude ($|M_{xy}|$) </p>
<p>There will be no signal at some off-resonance frequency, which will leads to banding artifact when reconstructe the image. </p>
<p>2. The relationship between TR, TE and the period of the off-resonance profile</p>


```python
TR = 0.01
TE = TR/2
T1 = 1
T2 = 0.1
tip_angle = [10, 30, 90]
#no phase cycling 

#----------------------------------------------------------------------------------------------
#draw for different tip angle, 10, 30, 90
plot_number = 1
for a in tip_angle:
    #store value
    off_resonance_f = np.asarray([], dtype = int)
    M_xy_steady = np.asarray([], dtype = float)
    M_phase_steady = np. asarray([], dtype = float)

    #Find steady state value for different off-resonance 
    for f_off in range(-400,405,5):
        M_pre = M0 #initial pre pulse magnetisation 
        #templorary store value
        M_xy = np.asarray([], dtype = float)
        M_phase = np.asarray([], dtype = float)

        #repeat 1000 times for each off-resonance frequency
        for k in range(1,TR_Num+1):
            #Tip
            M_post = RF_pulse(a,0) @ M_pre
            #Consider T1 recovery, T2 decay and precession
            #@ time TE 
            P_TE = Precession(f_off,TE) #precession matrix at t = TE
            E_TE = Relaxation(TE, T1, T2) #Relaxation matrix at t = TE

            M_TE = P_TE @ E_TE @ M_post + (np.identity(3) - E_TE) @ M0 #magnetisation at TE
            M_transverse = np.sqrt(M_TE[0]**2+M_TE[1]**2)
            phase = math.atan2(M_TE[1], M_TE[0]) / math.pi * 180
            
            #store value in the array
            M_xy = np.append(M_xy, M_transverse)
            M_phase = np.append(M_phase, phase)

            #@ time TR 
            P_TR = Precession(f_off,TR) #precession matrix at t = TR
            E_TR = Relaxation(TR, T1, T2) #Relaxation matrix at t = TR
            M_TR = P_TR @ E_TR @ M_post + (np.identity(3) - E_TR) @ M0 #magnetisation at TR
            M_pre = M_TR
        #END
     
        M_xy_steady = np.append(M_xy_steady, M_xy[-1])
        M_phase_steady = np.append(M_phase_steady, M_phase[-1])
        off_resonance_f = np.append(off_resonance_f, f_off)
    #END
    
    #plot M_xy at steady state against off-resonance frequency
    plt.figure(plot_number)
    fig, (ax1, ax2) = plt.subplots(2)
    fig.suptitle("tip angle:" + str(a))
    ax1.plot(off_resonance_f, M_xy_steady)
    ax1.set_xlabel("off resonance frequency (Hz)")
    ax1.set_ylabel("$|M_{xy}|$")
    ax1.set_xlim([-400, 400])
    ax1.grid(True)
    
    
    
    ax2.plot(off_resonance_f, M_phase_steady)
    ax2.set_xlabel("off resonance frequency (Hz)")
    ax2.set_ylabel("$Phase (degree)$")
    ax2.set_xlim([-400, 400])
    ax2.grid(True)
    plt.yticks(np.arange(-180, 181, 90))
    
    
    #plt.show()
    
    #save figure
    plt.savefig("PartTwo_" +str(int(T1*1000))+"_"+str(int(T2*1000))+"_TR" + str(TR) + "_off_resonance_profile_tip_angle_" + str(a) + ".jpg")
    
    plot_number = plot_number+1
```

## PART 3:
<p>For balance Steady State Free Precession with <b>phase cycling ($\Delta\phi = 90$)</b></p>
<p>Modeling the signals from spins with different off-resonance, <b>off-resonance range (-400 ~ 400 Hz)</b></p>
<p></p>

```python
TR = 0.005
TE = TR/2
T1 = 1
T2 = 0.1
phasecycle = [0, 90, 180] #[0, 45, 90, 135, 180]
tip_angle = [30]

#----------------------------------------------------------------------------------------------
#draw for different tip angle, 10, 30, 90
plot_number = 1
for a in tip_angle:
    
    for phase_cycle in phasecycle:
        #store value
        off_resonance_f = np.asarray([], dtype = int)
        M_xy_steady = np.asarray([], dtype = float)
        M_phase_steady = np. asarray([], dtype = float)


        #Find steady state value for different off-resonance 
        for f_off in range(-400,405,5):
            M_pre = M0 #initial pre pulse magnetisation 
            #templorary store value
            M_xy = np.asarray([], dtype = float)
            M_phase = np.asarray([], dtype = float)

            phase_shift = 0

            #repeat 1000 times for each off-resonance frequency
            for k in range(1,TR_Num+1):
                #Tip

                M_post = RF_pulse(a,phase_shift) @ M_pre
                #Consider T1 recovery, T2 decay and precession
                #@ time TE 
                P_TE = Precession(f_off,TE) #precession matrix at t = TE
                E_TE = Relaxation(TE, T1, T2) #Relaxation matrix at t = TE

                M_TE = P_TE @ E_TE @ M_post + (np.identity(3) - E_TE) @ M0 #magnetisation at TE
                M_transverse = np.sqrt(M_TE[0]**2+M_TE[1]**2)
                phase = math.atan2(M_TE[1], M_TE[0]) / math.pi * 180

                #store value in the array
                M_xy = np.append(M_xy, M_transverse)
                M_phase = np.append(M_phase, phase)


                #@ time TR 
                P_TR = Precession(f_off,TR) #precession matrix at t = TR
                E_TR = Relaxation(TR, T1, T2) #Relaxation matrix at t = TR
                M_TR = P_TR @ E_TR @ M_post + (np.identity(3) - E_TR) @ M0 #magnetisation at TR
                M_pre = M_TR
                phase_shift = phase_shift + phase_cycle
            #END
            
            #Grab and store the steady state value
            M_xy_steady = np.append(M_xy_steady, M_xy[-1])
            M_phase_steady = np.append(M_phase_steady, M_phase[-1])
            off_resonance_f = np.append(off_resonance_f, f_off)
        #END

        #plot M_xy at steady state against off-resonance frequency
        # plt.figure(plot_number)
        # fig, (ax1, ax2) = plt.subplots(2)
        # fig.suptitle("Off-Resonance Profile ($\Delta\phi = $" + str(phase_cycle)+") ")
        # ax1.plot(off_resonance_f, M_xy_steady)
        # ax1.set_xlabel("off resonance frequency (Hz)")
        # ax1.set_ylabel("$|M_{xy}|$")
        # ax1.set_xlim([-400, 400])
        # ax1.grid(True)



        #ax2.plot(off_resonance_f, M_phase_steady)
        #ax2.set_xlabel("off resonance frequency (Hz)")
        #ax2.set_ylabel("$Phase (degree)$")
        #ax2.set_xlim([-400, 400])
        #ax2.grid(True)
        #plt.yticks(np.arange(-180, 181, 90))
        
        #save figure
        #plot M_xy at steady state against off-resonance frequency
        
        #fig, (ax1, ax2) = plt.subplots(2)
        #fig.suptitle("Off-Resonance Profile ($\Delta\phi = $" + str(phase_cycle)+") ")
        plt.plot(off_resonance_f, M_xy_steady,label = "$\Delta\phi = $" + str(phase_cycle)+"")
        plt.xlabel("off resonance frequency (Hz)")
        plt.ylabel("$|M_{xy}|$")
        plt.ylim([0,0.2])
        plt.xlim([-400, 400])
        plt.grid(True)
        plt.legend(loc="upper right")
        #plt.savefig("PartThree_phasecycle"+str(phase_cycle)+"_TR_" + str(TR) + "_off_resonance_profile_tip_angle_" + str(a) + ".jpg")
plt.savefig("phase_cycling_bSSFP_planningrepo.jpg")
plt.show()
        #plot_number = plot_number+1
    #END
    
   
   
```

## PART 4:
<p>Periodic phase cycling, with $\Delta \phi_1 = 0, \Delta \phi_2 = 90$</p>


```python
T1 = 170e-3
T2 = 40e-3
TR = 4.4e-3
TE = TR/2
phase_cycle_1 = 0
phase_cycle_2 = 180
tip_angle = [30]

#----------------------------------------------------------------------------------------------
#draw for different tip angle, 10, 30, 90

for a in tip_angle:
    
    #store value
    off_resonance_f = np.asarray([], dtype = int)
   
    M_xy_steady_odd = np.asarray([], dtype = float)
    M_xy_steady_even = np.asarray([], dtype = float)

    #Find steady state value for different off-resonance 
    for f_off in range(-400,405,5):
        M_pre = M0 #initial pre pulse magnetisation 
        #templorary store value
        M_xy_odd = np.asarray([], dtype = float)
        M_xy_even = np.asarray([], dtype = float)
        phase_shift = 0

        #repeat 1000 times for each off-resonance frequency
        for k in range(1,150):
            #Tip
            M_post = RF_pulse(a,phase_shift) @ M_pre
            
            #Consider T1 recovery, T2 decay and precession
            #@ time TE 
            P_TE = Precession(f_off,TE) #precession matrix at t = TE
            E_TE = Relaxation(TE, T1, T2) #Relaxation matrix at t = TE

            M_TE = P_TE @ E_TE @ M_post + (np.identity(3) - E_TE) @ M0 #magnetisation at TE
            M_transverse = np.sqrt(M_TE[0]**2+M_TE[1]**2)

            #@ time TR 
            P_TR = Precession(f_off,TR) #precession matrix at t = TR
            E_TR = Relaxation(TR, T1, T2) #Relaxation matrix at t = TR
            M_TR = P_TR @ E_TR @ M_post + (np.identity(3) - E_TR) @ M0 #magnetisation at TR
            M_pre = M_TR
            
            if (k%2)==0: 
                #store value in the array
                M_xy_even = np.append(M_xy_even, M_transverse)
                #next shift
                phase_shift = phase_shift + phase_cycle_1
            else:
                #store value in the array
                M_xy_odd = np.append(M_xy_odd, M_transverse)
                #next shift
                phase_shift = phase_shift + phase_cycle_2
        #END
        M_xy_steady_odd = np.append(M_xy_steady_odd, M_xy_odd[-1])
        M_xy_steady_even = np.append(M_xy_steady_even, M_xy_even[-1])
        off_resonance_f = np.append(off_resonance_f, f_off)
    #END
    
    #plot M_xy at steady state against off-resonance frequency
    plt.figure(1)
    plt.plot(off_resonance_f, M_xy_steady_odd)
    plt.xlabel("off resonance frequency (Hz)")
    plt.ylabel("$|M_{xy}|$")
    plt.title("Off-resonance Profile (odd) tip angle:" + str(a))
    
    plt.figure(2)
    plt.plot(off_resonance_f, M_xy_steady_even)
    plt.xlabel("off resonance frequency (Hz)")
    plt.ylabel("$|M_{xy}|$")
    plt.title("Off-resonance Profile (even) tip angle:" + str(a))
    plt.savefig("FEMR.jpg")
    plt.show()
    
    #save figure
    
   
```

## PART 5 
<p>try PLANET on phase cycling simulation value <p/>
<p>for phasecycle $\Delta\phi = 45°$. Will store 8 points, one from each off-resonance plot at same off-resonance frequency<p/>

```python
TR = 0.01
TE = TR/2
T1 = 1
T2 = 0.1
phasecycle = [0, 45, 90, 135, 180, 225, 270, 315]
tip_angle = [30]

#----------------------------------------------------------------------------------------------
#draw for different tip angle, 10, 30, 90
plot_number = 1
for a in tip_angle:
    #at same resonance frequency, store the Mx and My value at different phase-cycling value. Get 8 points. 
    planet_data_x_1 = np.asarray([], dtype = float)
    planet_data_y_1 = np.asarray([], dtype = float)
    planet_data_x_2 = np.asarray([], dtype = float)
    planet_data_y_2 = np.asarray([], dtype = float)
    planet_data_x_3 = np.asarray([], dtype = float)
    planet_data_y_3 = np.asarray([], dtype = float)
    
    for phase_cycle in phasecycle:
        #store value
        off_resonance_f = np.asarray([], dtype = int)
        #M_xy_all = np.asarray([], dtype = float)
        M_xy_steady = np.asarray([], dtype = float)


        #Find steady state value for different off-resonance 
        for f_off in range(-400,405,5):
            M_pre = M0 #initial pre pulse magnetisation 
            #templorary store value
            M_xy = np.asarray([], dtype = float)

            phase_shift = 0

            #repeat 1000 times for each off-resonance frequency
            for k in range(1,TR_Num+1):
                #Tip

                M_post = RF_pulse(a,phase_shift) @ M_pre
                #Consider T1 recovery, T2 decay and precession
                #@ time TE 
                P_TE = Precession(f_off,TE) #precession matrix at t = TE
                E_TE = Relaxation(TE, T1, T2) #Relaxation matrix at t = TE

                M_TE = P_TE @ E_TE @ M_post + (np.identity(3) - E_TE) @ M0 #magnetisation at TE
                M_transverse = np.sqrt(M_TE[0]**2+M_TE[1]**2)

                #store value in the array
                M_xy = np.append(M_xy, M_transverse)

                #@ time TR 
                P_TR = Precession(f_off,TR) #precession matrix at t = TR
                E_TR = Relaxation(TR, T1, T2) #Relaxation matrix at t = TR
                M_TR = P_TR @ E_TR @ M_post + (np.identity(3) - E_TR) @ M0 #magnetisation at TR
                M_pre = M_TR
                phase_shift = phase_shift + phase_cycle
                if ((f_off == 0)&(k==200)):
                    planet_data_x_1 = np.append(planet_data_x_1, M_TE[0])
                    planet_data_y_1 = np.append(planet_data_y_1, M_TE[1])
                if ((f_off == 200)&(k==200)):
                    planet_data_x_2 = np.append(planet_data_x_2, M_TE[0])
                    planet_data_y_2 = np.append(planet_data_y_2, M_TE[1])
                if ((f_off == 400)&(k==200)):
                    planet_data_x_3 = np.append(planet_data_x_3, M_TE[0])
                    planet_data_y_3 = np.append(planet_data_y_3, M_TE[1])
            
             
            #END 
            
            M_xy_steady = np.append(M_xy_steady, M_xy[-1])

            off_resonance_f = np.append(off_resonance_f, f_off)
        #END

        #plot M_xy at steady state against off-resonance frequency
        plt.figure(plot_number)

        plt.plot(off_resonance_f, M_xy_steady)
        plt.xlabel("off resonance frequency (Hz)")
        plt.ylabel("$|M_{xy}|$")
        plt.title("Off-Resonance Profile ($\Delta\phi = $" + str(phase_cycle)+") ")
        

        #save figure
        #plt.savefig("PartThree_phasecycle"+str(phase_cycle)+"_TR_" + str(TR) + "_off_resonance_profile_tip_angle_" + str(a) + ".jpg")
        plt.show()
        plot_number = plot_number+1
    #END
    plt.figure(plot_number)
    plt.scatter(planet_data_x_1,planet_data_y_1)
    plt.scatter(planet_data_x_2,planet_data_y_2)
    plt.scatter(planet_data_x_3,planet_data_y_3)
   
    plt.show()
   
   
```

```python
points = np.column_stack((planet_data_x_1,planet_data_y_1))
x = points[:, 0]
y = points[:, 1]

from skimage.measure import EllipseModel
from matplotlib.patches import Ellipse

ell = EllipseModel()
ell.estimate(points)
xc, yc, a, b, theta = ell.params
print("center = ",  (xc, yc))
print("angle of rotation = ",  theta)
print("axes = ", (a,b))

fig, axs = plt.subplots(2, 1, sharex=True, sharey=True)
axs[0].scatter(x,y)
axs[1].scatter(x, y)
ell_patch = Ellipse((xc, yc), 2*a, 2*b, theta*180/np.pi, edgecolor='red', facecolor='none')
axs[1].add_patch(ell_patch)
plt.show()
```
